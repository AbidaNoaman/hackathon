```python
e_c_train_data = pd.read_csv(r'train_data.csv')
df = e_c_train_data
df
```

```python
# Rename the column Tweet to Prompt
```

```python
df.rename(columns={'Tweet': 'prompt'}, inplace=True)
prompt = df['prompt']
df.head()
```

```python
# Join all the output label via seperated by comma
```

```python
# Merge Multi-labels Output into a single column seperated by Commas and rename column as completion

outputs = df.iloc[:,1:]
outputs["final"] = outputs.apply(lambda row: row[row == 1].index, axis=1)
final_output = outputs["final"].to_frame()
final_output['final'] = final_output['final'].astype(str).str.replace("[\]\[]",'').astype(str)


new_data = []
for i in range (len(final_output)):
    new_data.append(final_output['final'][i][:-17])
new_data2 = []
for i in range (len(final_output)):
    new_data2.append(new_data[i][6:])  
    
   
completion = pd.DataFrame(new_data2,columns = ["completion"])
completion['completion'] = completion['completion'].str.replace(r'"', '')

#Join the Input prompt and output completion
prompt = prompt.to_frame()
emotions_dataset = prompt.join(completion)
emotions_dataset
```

```python
# Choose the subset
```

```python
emotions_dataset = emotions_dataset.iloc[:100,:]
emotions_dataset["completion"]=emotions_dataset["completion"].str.replace(' ','')
emotions_dataset["completion"]=emotions_dataset["completion"].str.replace("'",'')
emotions_dataset.head()
```

```python
emotions_dataset.to_json("emotions_dataset.jsonl", orient='records', lines=True)
```

```python
def upload_file(api_key, file_path):
    headers = {'Authorization': f'Bearer {api_key}'}
    files = {'purpose': (None, 'fine-tune'), 'file': open(file_path, 'rb')}
    response = requests.post('https://api.openai.com/v1/files', headers=headers, files=files)
    return response.json()

def create_fine_tuning_job(api_key, training_file_id):
    headers = {'Content-Type': 'application/json', 'Authorization': f'Bearer {api_key}'}
    data = {'training_file': training_file_id, 'model': 'babbage-002', 'hyperparameters': {'n_epochs': 1}}
    response = requests.post('https://api.openai.com/v1/fine_tuning/jobs', headers=headers, data=json.dumps(data))
    return response.json()

def get_fine_tuning_jobs(api_key, limit=1):
    url = "https://api.openai.com/v1/fine_tuning/jobs"
    params = {"limit": limit}
    headers = {"Authorization": f"Bearer {api_key}"}
    response = requests.get(url, params=params, headers=headers)
    return response.json()

def generate_emotion(api_key, fine_tuned_model_id, text):
    headers = {'Content-Type': 'application/json', 'Authorization': f'Bearer {api_key}'}
    data = {
        'model': fine_tuned_model_id,
        'prompt': f'You are requested to Please generate only emotions that are provided in the dataset: below is the Prompt:\n{text} ->',
        'max_tokens': 20,
        'stop': '\n'
    }
    response = requests.post('https://api.openai.com/v1/completions', headers=headers, data=json.dumps(data))
    return response.json()["choices"][0]['text']

```

```python
api_key = ''
```

```python
file_path = 'emotions_dataset_prepared (2).jsonl'

file_upload_response = upload_file(api_key, file_path)

print("file-id : ",file_upload_response['id'])
print("status : ",file_upload_response['status'])
```

```python
training_file_id = file_upload_response.get('id', '')

if training_file_id:
    fine_tuning_job_response = create_fine_tuning_job(api_key, training_file_id)
    print('job-id : ',fine_tuning_job_response['id'])
```

```python
response_text = get_fine_tuning_jobs(api_key)

model = response_text["data"][0]['fine_tuned_model']

print("Status Code :", response_text['data'][0]['status'])
print("Fine-tuned Model :", response_text["data"][0]['fine_tuned_model'])
```

```python
fine_tuned_model_id = model

text_to_generate_emotion = input("enter the text: ")
prompt = f"""You are request to please generate the emotions that are only available in the dataset. The prompt is as follows:
             Prompt: {text_to_generate_emotion} ->"""

generated_emotion = generate_emotion(api_key, fine_tuned_model_id, text_to_generate_emotion)

print("Emotions are; ",generated_emotion)
```
